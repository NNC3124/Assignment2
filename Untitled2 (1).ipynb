2.From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.
There are multiple ways of doing this. You can use pyenchant (https://pypi.org/project/pyenchant/), pyspellchecker (https://pyspellchecker.readthedocs.io/en/latest/) or just download a list of words (http://www.gwicks.net/dictionaries.htm) and search through them.


```python
from spellchecker import SpellChecker
import re

# Initialize the spell checker
spell = SpellChecker()

# Define a regular expression pattern to match words
WORD_RE = re.compile(r'\w+')

# Read the text from the file using UTF-8 encoding
with open('file2.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Define a function to preprocess words (remove punctuation)
def preprocess_word(word):
    return re.sub(r'[^\w\s]', '', word)  # Removes all non-alphanumeric characters except spaces

# Split text into words
words = text.split()

# Define a mapper function to emit non-English words
def mapper(word):
    word = preprocess_word(word)
    correction = spell.correction(word)
    if correction != word:
        return (word, 1)
    return (word, 0)

# Define a reducer function to count non-English words
def reducer(counts, word_count):
    word, count = word_count
    counts[word] += count
    return counts

# Map and reduce to get counts of non-English words
non_english_counts = reduce(reducer, map(mapper, words), Counter())

# Print non-English words and their counts
for word, count in non_english_counts.items():
    if count > 0:
        print(f'{word}: {count}')
```
